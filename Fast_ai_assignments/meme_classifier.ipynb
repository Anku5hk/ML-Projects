```
from fastai.vision import *
```

```
%reload_ext autoreload
%autoreload 2
%matplotlib inline
```

```
path = "/content/drive/My Drive/path/"
```

```
data = ImageDataBunch.from_folder(path, size=299, bs=5).normalize(imagenet_stats)
```

```
data.show_batch(5)
```

```
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```
learn.fit_one_cycle(5)
```

```
learn.lr_find()
```

```
learn.recorder.plot()
```

```
model_interpreter = ClassificationInterpretation.from_learner(learn)
```

```
model_interpreter.plot_confusion_matrix(figsize=(10,15))
```
