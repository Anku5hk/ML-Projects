# Custom filter

```python
# derivative of image in x direction

kernel = np.array([[-1,0,1],[-1,0,1],[-1,0,1]]) / 3
fil_img = cv2.filter2D(cvimg, -1, kernel)
plt.imshow(fil_img)
plt.show()
```

```python
# derivative of image in y direction

kernel = np.array([[1,1,1],[0,0,0],[-1,-1,-1]]) / 3
fil_img = cv2.filter2D(cvimg, -1, kernel)
plt.imshow(fil_img)
plt.show()
```

# Blur image

```python
# blur image filter(replace each pixel by average of neighbour)

kernel = np.zeros((3,3))
kernel = kernel + 1 / 9
filt_img = cv2.filter2D(img, -1,kernel) # same as cv2.GuassianBlur
plt.imshow(Image.fromarray(filt_img))
plt.show()
```

# Image sharpening filter

```python
img = np.asarray(img)

kernel1 = np.zeros((3,3), dtype=int)
kernel1[1][1] = 2
kernel2 = np.zeros((3,3), dtype=int)
image_shaprnening_filter = kernel1 - (kernel2 + 1)/ 9

fil_img = cv2.filter2D(img, -1, image_shaprnening_filter)
plt.imshow(Image.fromarray(fil_img))
plt.show()
```

# Median filtering(non linear filter)

```python
median = cv2.medianBlur(img, 5)
plt.imshow(median)
plt.show()
```

# Add/Remove noise

```python
noise = np.zeros((400,400,4))
noise = cv2.randn(noise,(0),(100))
noise_img = img + noise
plt.imshow(np.clip(noise_img,1,255).astype(int))
plt.title("Added Noise")
plt.show()
```

```python
# add noise
noise = np.zeros((400,400,3))
noise = cv2.randn(noise,(0),(100)) # norm dist
noise_img = cvimg + noise
plt.imshow(np.clip(noise_img,0,255).astype(int))
plt.show()
```

```python
# remove noise 

blr_img = cv2.GaussianBlur(noise_img, (5,5), 0)
plt.imshow(np.clip(blr_img,0,255).astype(int))
plt.show()
```
